```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv')
parent_theme = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv')
```

```python
df.head()
```

```python
parent_theme.head()
```

```python
merged = df.merge(parent_theme, left_on="parent_theme", right_on="name")
merged
```

```python
merged.drop(["name_y"], axis=1, inplace=True)
merged
```

```python
#drop set_num null rows
merged[merged["set_num"].isnull()].shape[0]
```

Licensed Star Wars Percentage

```python
licensed = merged[merged["is_licensed"]]
licensed = licensed.dropna(subset="set_num")

licensed_star_wars = licensed[licensed["parent_theme"] == "Star Wars"]

percentage = (licensed_star_wars.shape[0]/licensed.shape[0]*100)
percentage
```

Find not most popular year for Star Wars

```python
licensed["count"] = 1

summed_df = licensed.groupby(["year","parent_theme"]).sum().reset_index()
max_df = summed_df.sort_values("count", ascending=False).drop_duplicates(['year'])

max_df.sort_values("year", ascending=False, inplace=True)
```

```python
new_era = max_df[max_df["parent_theme"] != "Star Wars"]
new_era
```

```python
new_era[["year", "parent_theme", "count"]]
```

Break down number of sets by year

```python
clean_df = merged.dropna(subset="set_num")
clean_df["count"] = 1
sumofset_by_year = clean_df.groupby("year").sum().reset_index()[["year", "count"]]

sumofset_by_year
```
